```python
import numpy as np
import matplotlib.pyplot as plt

from src import plummer
from src import leapint
```

```python
# G = 1
M_unit = 10e7 # M_solar = 1.99e41 kg
r_unit = 1.5 # Kpc
t_unit = 2.7327 # 10e8year

N = int(1e4)
steps = 150
```

```python
x_stable = np.empty((steps, N, 3))
v_stable = np.empty_like(x_stable)


with open("plummer.tab", "r") as f:

    for i in range(steps):
        f.readline()

        for j in range(N):
            line = f.readline().split()

            for k in range(3):
                x_stable[i, j, k] = float(line[k+1])
                v_stable[i, j, k] = float(line[k+4])
```

```python
E_hist_stable = np.array(tuple(np.sum(plummer.E(x_stable[i], v_stable[i])) for i in range(steps)))
```
